9. Calculate the Average Final Grade

The final examination score is stored in the **G3** column.

Using the `mean()` function, we calculate the average performance of all students.

This helps evaluate the overall academic performance in the dataset.

Average Final Grade (G3)
The average final grade (G3) of students is around 10–11 marks.
This indicates that most students achieved a moderate academic performance.

```python
above15 = df[df["G3"] > 15]

print("Students scoring above 15:", len(above15))
```

 10. Students Scoring Above 15

Students with final grades greater than 15 are identified using filtering.

This analysis helps determine how many students achieved excellent academic performance.

Students Scoring Above 15
Only a small proportion of students scored above 15, showing that high achievers are relatively few.
Most students scored between 8 and 14 marks.

```python
correlation = df["studytime"].corr(df["G3"])

print("Correlation:", correlation)
```

 11. Relationship Between Study Time and Final Grades

The correlation coefficient is calculated between:

- Study Time (`studytime`)
- Final Grade (`G3`)

Correlation values range from **-1 to +1**.

- Positive values indicate a positive relationship.
- Negative values indicate a negative relationship.
- Values close to zero indicate little or no relationship.

This helps determine whether increased study time improves student performance.

Study Time vs Performance:
Students who spend more time studying generally tend to obtain higher grades, indicating a positive correlation.

```python
gender_average = df.groupby("sex")["G3"].mean()

print(gender_average)
```

12. Compare Performance by Gender

The average final grades are calculated separately for male and female students.

This comparison helps identify whether one group performed better on average.

The `groupby()` function is used for this analysis.

Gender-wise Performance:
Female students have a slightly higher average final grade compared to male students, though the difference is not substantial.

```python
plt.figure(figsize=(8,5))

plt.hist(df["G3"], bins=10)

plt.title("Distribution of Final Grades")

plt.xlabel("Final Grade")

plt.ylabel("Number of Students")

plt.show()
```

13. Distribution of Final Grades

A histogram is used to visualize the distribution of final grades.

It shows:

- Most common grades
- Spread of marks
- Frequency of student scores

This visualization helps understand the overall performance of students.

```python
plt.figure(figsize=(8,5))

plt.scatter(df["studytime"], df["G3"])

plt.title("Study Time vs Final Grade")

plt.xlabel("Study Time")

plt.ylabel("Final Grade")

plt.show()
```

14. Study Time vs Final Grade

A scatter plot is created to examine the relationship between study time and final grades.

Each point represents one student.

This graph helps identify whether students who study more generally achieve higher marks.

```python
gender_average.plot(kind="bar")

plt.title("Average Final Grade by Gender")

plt.xlabel("Gender")

plt.ylabel("Average Grade")

plt.show()
```

15. Average Final Grade by Gender

A bar chart compares the average final grades of male and female students.

The height of each bar represents the average score.

This visualization makes gender-based performance comparisons easy to understand.
